# 지역별 업종별 긍정 게시량 2019년, 2020년 비교

## 1. UP_POS 그래프 함수 만들기

```python
def region_up_pos(region, loc, tit):
    index = np.arange(4)
    ls = ["2월", "3월", "4월", "5월"]
    bar_width = 0.35
    alpha = 0.5
    
    plt.figure(figsize=(20, 10))
    
    p1 = plt.bar(index, loc[0:4], bar_width, color='b', alpha=alpha, label="2019년")
    p2 = plt.bar(index + bar_width, loc[4:8], bar_width, color='r', alpha=alpha, label="2020년")

    plt.title(tit, fontsize=20)
    plt.ylabel('Count', fontsize=18)
    plt.xlabel('Date', fontsize=18)
    plt.xticks(index, ls, fontsize=15)
    plt.legend((p1[0], p2[0]), ("2019년", "2020년"), fontsize=15)
    plt.savefig(tit + '.png', bbox_inches='tight')
    
    plt.show()
```
