## 환경 부담금 지불 할 의사 Data

- 출처 : <http://data.seoul.go.kr/dataList/OA-15565/F/1/datasetView.do>

```python
survey = pd.read_excel("4. 2018 서베이 가구원(원시자료와 부호화 지침서).xlsx")
```

```python
survey
```

### 필요한 컬럼만 뽑아주기

```python
survey_df = survey[["DE6", "AQ10"]].rename(columns={"DE6":"분류", "AQ10":"willing_to_pay"})
```

```python
# 행정구역 코드를 알아보기 쉽게 한글명으로 변경
change_value = {    
    110 : "종로구",
    140 : "중구",
    170 : "용산구",
    200 : "성동구",
    215 : "광진구",
    230 : "동대문구",
    260 : "중랑구",
    290 : "성북구",
    305 : "강북구",
    320 : "도봉구",
    350 : "노원구",
    380 : "은평구",
    410 : "서대문구",
    440 : '마포구',
    470 : "양천구",
    500 : "강서구",
    530 : "구로구",
    545 : "금천구",
    560 : "영등포구",
    590 : "동작구",
    620 : '관악구',
    650 : "서초구",
    680 : "강남구",
    710 : "송파구",
    740 : "강동구"
}

survey_df = survey_df.replace({"분류":change_value})
```

### 각 지역별추가 지출 의사 

```python
willing_df = survey_df.groupby("분류", as_index = False).mean()
```

```python
# willing_to_pay 높은 순으로 뽑기
willing_df.sort_values("willing_to_pay", ascending = False)
```

## 생활 에너지 절약 Data

### 필요한 데이터 전처리

```python
save = pd.read_csv("201_DT_201_10550_2015_20200912162344.csv", encoding='cp949')
save
```

```python
# 구별로 처리할 데이터만 분리
save_df = save[:150]
save_df.drop(columns=["대분류", "항목", "단위", "Unnamed: 6"], inplace=True)
```

```python
# 구별로 항목 컬럼 생성
save_df = save_df.pivot_table(index="분류", columns="구분", values="2019 년")
```

```python
# 평균값도 하나 뽑아주자
save_df["평균"] = save_df.mean(axis=1)
```

```python
save_df.reset_index(drop=False, inplace=True)
```

## 데이터 셋 하나로 합쳐주기

```python
eco_pay = pd.merge(save_df, willing_df, on = "분류")
eco_pay
```
