```python
X = vectorizer.fit_transform(mydf['키워드'])
```

```python
content = mydf['키워드']

mycontent = content.apply(lambda row: row.split(','))
```

```python
dictionary = gensim.corpora.Dictionary(mycontent)

dictionary.filter_tokens()

min_count=5

word_counter = Counter((word for words in mycontent for word in words))

removal_word_idxs = {
    dictionary.token2id[word] for word, count in word_counter.items()
    if count < min_count
}
```

```python
class Corpus:
    def __init__(self, path, dictionary):
        self.path = path
        self.dictionary = dictionary
        self.length = 0
    def __iter__(self):
        for doc in self.path:
            yield self.dictionary.doc2bow(doc)
    def __len__(self):
        if self.length == 0:
            for i, doc in enumerate(self.path):
                continue
            self.length = i + 1
        return self.length
```

```python
corpus = Corpus(mycontent,dictionary)

from gensim.models import LdaModel

lda_model = LdaModel(corpus, id2word=dictionary, num_topics=50)

import pyLDAvis.gensim as gensimvis

prepared_data = gensimvis.prepare(lda_model, corpus, dictionary)

dictionary.filter_tokens(removal_word_idxs)
dictionary.compactify()

import pyLDAvis
```

```python
pyLDAvis.display(prepared_data)
```

## 코로나 기사 키워드 빈도

```python
content = mydf['키워드']

mycontent = content.apply(lambda row: row.split(','))
```

```python
mydf['일자'] = mydf['일자'].apply(lambda x: str(x))
```

```python
mydf_02 = mydf[mydf['일자'].str.startswith('202002')]
mydf_03 = mydf[mydf['일자'].str.startswith('202003')]
mydf_04 = mydf[mydf['일자'].str.startswith('202004')]
mydf_05 = mydf[mydf['일자'].str.startswith('202005')]
mydf_06 = mydf[mydf['일자'].str.startswith('202006')]
```

```python
mydfs = [mydf_02, mydf_03, mydf_04, mydf_05, mydf_06]
```

```python
def find_amount_of_keyword(query):
    for i in range(len(mydfs)):
        temp = mydfs[i]['키워드'].apply(lambda x: x.split(','))
        temp2 = []
        for a in temp:
            temp2.extend(a)
        temp3 = [x for x in temp2 if x==query]
        print('2020년 '+str(i+2)+' 월의 '+query+' 키워드 개수는 '+str(len(temp3))+' 입니다.')
```

```python
find_amount_of_keyword('배달')
```
