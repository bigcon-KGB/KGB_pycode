### 3) 2014-2017 년도 생활 폐기물 발생량 data 생성 

#### 2014-2017 년도 생활 폐기물 발생량 data 생성하는 class 

```python
class Trash_data_modify:
    ## csv 파일 read 하기 (생성자 함수)
    def __init__(self, year):
        self.original_data = pd.read_excel("04_02_{}_전국_시군구 현황.xlsx".format(year), sheet_name="2가+나", header=5)
        self.total_data = self.original_data.copy()
        self.total_data = self.total_data[self.total_data["Unnamed: 2"]=="발생량"]
        
        self.column = pd.read_excel("04_02_{}_전국_시군구 현황.xlsx".format(year), sheet_name="2가+나", header=4, nrows=1)
    
    
    def column_change(self, num1, num2):
        ### column 정리01_trash_data_calumn 의 column 일부를 trash_amount_data로 변경
        for before_column, after_column in zip(self.total_data.columns[num1:num2], self.column.columns[num1:num2]):
            self.total_data.rename(columns={before_column:after_column}, inplace=True)
    
        ### trash_amount_data의 column 정리 02
        self.total_data.rename(columns={"Unnamed: 4":"발생량_총계", 
                                        "Unnamed: 5":"종량제_총계", "Unnamed: {}".format(num2-1):"음식물류_폐기물_분리배출"}, inplace=True)
        self.total_data.drop(columns=["Unnamed: 2", "Unnamed: 3"], inplace=True)
        self.total_data = self.total_data.reset_index(drop=True)

        ### column 정리 03_종량제, 재활용 구분
        for n in range(4,11):
            self.total_data.columns.values[n] = '종량제_가연성_'+ self.total_data.columns.values[n]
        for n in range(11,16):
            self.total_data.columns.values[n] = '종량제_불연성_'+ self.total_data.columns.values[n]
        if num2 >= 37:
            self.total_data.columns.values[16] = "종량제_기타(배출불명등)"
            for n in range(17, num2-3):
                self.total_data.columns.values[n] = '재활용_'+ self.total_data.columns.values[n]

        else:
            for n in range(16, num2-3):
                self.total_data.columns.values[n] = '재활용_'+ self.total_data.columns.values[n]
            
        return self.total_data

    
    ## total data를 csv로 저장하기
    def save_total_data_csv(self, year):
        self.total_data.to_csv("{}_Total_Trash_Amount_Data.csv".format(year), index = False)
    
    
    ## 필요 데이터만 뽑아 정리
    def final_data(self):
        ### 발생량_총계 / 종량제, 재활용 의 총계 및 플라스틱 발생량만 뽑아서 데이터 생성
        self.data = self.total_data[["시도", "시군구", "발생량_총계", "종량제_총계", 
                                                    "종량제_가연성_플라\n스틱류", "재활용_소계.1", "재활용_플라\n스틱류", "음식물류_폐기물_분리배출"]]
        ### 위에서 뽑은 데이터 중 전국, 서울, 대구만 뽑아서 데이터 정리 및 column 최종 정리
        all_country = self.data["시도"]=="전국"
        seoul = self.data["시도"]=="서울"
        daegu = self.data["시도"]=="대구"
        self.data = self.data[all_country | seoul | daegu]
        self.data.rename(columns={"종량제_가연성_플라\n스틱류":"종량제_가연성_플라스틱류", 
                                          "재활용_소계.1":"재활용_총계",
                                          "재활용_플라\n스틱류":"재활용_플라스틱류"}, inplace = True)
        self.data["지역별_플라스틱_총계"] = self.data["종량제_가연성_플라스틱류"] + self.data["재활용_플라스틱류"]
        self.data["지역별_플라스틱 비중(%)"] = self.data["지역별_플라스틱_총계"] / self.data["발생량_총계"] * 100

        self.data = self.data.reset_index(drop=True)
        return self.data
    
    
    ## final data를 csv로 저장하기
    def save_final_data_csv(self, year):
        self.data.to_csv("{}_Final_Trash_Amount_Data.csv".format(year), index = False)
```

## 2. 2014-2018 생활 폐기물 지역별 처리현황 data

#### 2014-2018 년도 생활 폐기물 처리현황 data 생성하는 class

```python
class Trash_disposal_data:
    ## csv 파일 read 하기 (생성자 함수)
    def __init__(self, year, sheet="3가+나"):
        self.original_disposal = pd.read_excel("04_02_{}_전국_시군구 현황.xlsx".format(year), sheet_name=sheet, header=4, usecols="A:F")
        self.original_disposal["시도"].fillna(method='ffill', inplace=True)
                
    
    ## original_disposal_data를 csv로 저장
    def save_original_disposal_data(self, year):
        self.original_disposal.to_csv("{}_Total_Trash_Disposal_Data.csv".format(year), index = False)
    
    
    ## final_disposal_data 생성
    def final_disposal_data(self):
        all_country = self.original_disposal["시도"]=="전국"
        seoul = self.original_disposal["시도"]=="서울"
        daegu = self.original_disposal["시도"]=="대구"
        self.original_disposal = self.original_disposal[all_country | seoul | daegu]
        self.final_disposal = self.original_disposal.reset_index(drop=True)
        
        self.final_disposal.rename(columns={self.final_disposal.columns[2]:"처리량_합계"}, inplace=True)
        
        self.final_disposal["재활용 비중(%)"] = self.final_disposal["재활용"] / self.final_disposal["처리량_합계"] * 100
        
        return self.final_disposal
    
    
    ## final_disposal_data를 csv로 저장
    def save_final_disposal_data(self,year):
        self.final_disposal.to_csv("{}_Final_Disposal_Data.csv".format(year), index = False)
```

## 3. 2014-2018 생활 폐기물 종류별 처리현황 data & 폐플라스틱 처리현황 data

#### 2014-2018 년도 생활 폐기물 종류별 및 폐플라스틱 처리현황 data 생성하는 class

```python
class Plastic_disposal:
    ## csv 파일 read 하기 (생성자 함수)
    def __init__(self, year):
        self.original_data = pd.read_excel("04_02_{}_전국_시군구 현황.xlsx".format(year), sheet_name="2가+나", header=5)    
        self.original_data.drop(columns="Unnamed: 2", inplace=True) # 발생량 데이터 삭제
        
        self.column = pd.read_excel("04_02_{}_전국_시군구 현황.xlsx".format(year), sheet_name="2가+나", header=4, nrows=1)
        
        ### 결측지 제거
        self.original_data["시도"].fillna(method='ffill', inplace=True)
        self.original_data["시군구"].fillna(method='ffill', inplace=True)
        self.original_data.fillna("처리량", inplace=True)
         
    
    def column_change(self, num1, num2):
        ### column 정리01_plastic_disposal_calumn 의 column 일부를 plastic_disposal_data로 변경
        if num2 >= 37:
            for before_column, after_column in zip(self.original_data.columns[num1-1:num2], self.column.columns[num1:num2]):
                self.original_data.rename(columns={before_column:after_column}, inplace=True)
        else:
            for before_column, after_column in zip(self.original_data.columns[num1-2:num2], self.column.columns[num1-1:num2]):
                self.original_data.rename(columns={before_column:after_column}, inplace=True)

        
        self.original_data.rename(columns={"Unnamed: 3":"처리방법", "Unnamed: 4":"처리량_총계", 
                                        "Unnamed: 5":"종량제_소계", "Unnamed: {}".format(num2):"음식물류_폐기물_분리배출"}, inplace=True)
        
        ### column 정리 02_종량제, 재활용 구분
        for n in range(5,12):
            self.original_data.columns.values[n] = '종량제_가연성_'+ self.original_data.columns.values[n]
        for n in range(12,17):
            self.original_data.columns.values[n] = '종량제_불연성_'+ self.original_data.columns.values[n]
        if num2 >= 37:
            self.original_data.columns.values[17] = "종량제_기타(배출불명등)"        
            for n in range(18, num2-1):
                self.original_data.columns.values[n] = '재활용_'+ self.original_data.columns.values[n]
        else:
            for n in range(17, num2-1):
                self.original_data.columns.values[n] = '재활용_'+ self.original_data.columns.values[n]
                
        return self.original_data
    
    
    ## original_plastic_data를 csv로 저장하기
    def save_original_plastic_disposal_csv(self, year):
        self.original_data.to_csv("{}_Total_Plastic_Disposal_Data.csv".format(year), index = False)
    
    
    ## plastic_disposal와 전국, 서울, 대구 만 뽑아서 data 생성
    def plastic(self):
        self.plastic = pd.concat([self.original_data.iloc[:,0:4], self.original_data.filter(like="플라",axis=1)], axis=1)        
        
        ### 위에서 뽑은 데이터 중 전국, 서울, 대구만 뽑아서 데이터 정리 및 column 최종 정리
        all_country = self.plastic["시도"]=="전국"
        seoul = self.plastic["시도"]=="서울"
        daegu = self.plastic["시도"]=="대구"
        self.plastic = self.plastic[all_country | seoul | daegu]
        self.plastic.rename(columns={"종량제_가연성_플라\n스틱류":"종량제_가연성_플라스틱류", 
                                          "재활용_플라\n스틱류":"재활용_플라스틱류"}, inplace = True)
        self.plastic = self.plastic.reset_index(drop=True)
        
        ### 처리현황별_플라스틱 비중(%) : 전체 생활폐기물 처리량 중 플라스틱의 처리현황 비중
        ### 플라스틱 재활용 비중(%) : 순수 재활용_플라스틱만의 비중
        self.plastic["처리현황별_플라스틱_총계"] = self.plastic["종량제_가연성_플라스틱류"] + self.plastic["재활용_플라스틱류"]
        self.plastic["처리현황별_플라스틱 비중(%)"] = self.plastic["처리현황별_플라스틱_총계"] / self.plastic["처리량_총계"] * 100
        self.plastic["플라스틱 재활용 비중(%)"] = self.plastic["재활용_플라스틱류"] / self.plastic["처리량_총계"] * 100


        return self.plastic
    
    
    ## plastic_disposal_data를 csv로 저장하기
    def save_final_plastic_disposal_csv(self, year):
        self.plastic.to_csv("{}_Final_Plastic_Disposal_Data.csv".format(year), index = False)
```

#### 2018 생활 폐기물 종류별 및  폐플라스틱 처리현황 data

```python
plastic_2018 = Plastic_disposal(2018)
```
